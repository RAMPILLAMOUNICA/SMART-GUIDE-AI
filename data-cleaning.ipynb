```python
import pandas as pd

data = pd.read_csv("laptop_pred.csv")

print(data.head())
print(data.columns)
print(data.shape)
```

```python
#data cleaning 
data = data.drop(['Unnamed: 0.1','Unnamed: 0'], axis=1)

print(data.head())
print(data.shape)
```

```python
#Check Missing Values
print(data.isnull().sum())
```

```python
#Check Data Types
print(data.info())
```

```python
#feature selection
features = data[['brand','price','spec_rating','processor','Ram','ROM','GPU']]

print(features.head())
```

```python
#text to numbers 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in ['brand','processor','Ram','ROM','GPU']:
    features[col] = le.fit_transform(features[col])

print(features.head())
```

```python
#laptop recommendation 
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=5, metric='euclidean')

model.fit(features)

print("Model trained successfully!")
```

```python
import numpy as np

# choose a laptop index (example: first laptop)
query = features.iloc[0].values.reshape(1, -1)

distances, indices = model.kneighbors(query)

print(indices)
```

```python
print(data.iloc[indices[0]])
```

```python
laptop_index = int(input("Enter laptop index: "))

query = features.iloc[laptop_index].values.reshape(1, -1)

distances, indices = model.kneighbors(query)

print(data.iloc[indices[0]])
```
